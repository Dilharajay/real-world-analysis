#Feature Engeneering

```python
import pandas as pd

data = {
    'Student_ID': [1, 2, 3, 4, 5],
    'Height_cm': [160, 170, 165, 180, 175],
    'Weight_kg': [55, 68, 60, 85, 78],
    'Hours_Studied': [4, 6, 5, 7, 3],
    'Attendance': [80, 90, 85, 95, 70],
    'Assignment_Score': [75, 85, 78, 92, 65],
    'Final_Grade': ['B', 'A', 'B', 'A', 'C']
}

df = pd.DataFrame(data)

print(df)
```

```python
df['Height_m'] = df['Height_cm']/100
df
```

```python
df['BMI'] = df['Weight_kg']/df['Height_m']**2
df
```

```python
df['Study_Efficiency'] = df['Hours_Studied'] * df['Attendance']
df
```

```python
df['Performance_Ratio'] = df['Assignment_Score'] / df['Hours_Studied']
df
```

**Original Features:**

* Student-ID
* Height_cm
* Weight_kg
* Hours_Studied
* Attendance
* Assignment_Score
* Final_Grade

**Derived Features**

* Height_m
* BMI
* Study_Efficiency
* Performance_Ratio

```python
corrmatrix = df.corr(numeric_only=True)
corrmatrix
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8,8))
plt.title('Correlation Heatmap')
sns.heatmap(corrmatrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.show()
```
