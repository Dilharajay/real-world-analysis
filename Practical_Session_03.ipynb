#Part A - Setup

```python
import pandas as pd
import numpy as np
```

```python
data = {
    "Student_ID" : [1,2,3,4,5],
    "Name" : ["Alice", "Bob", "Charlie", "David", "Eve"],
    "Attendence" : [85, 92, 78, 95, 88],
    "Final_Grade" : ["A", "B", "C", "A", "B"],
    "Enrolled_Date" : ["2023-07-01", "2023-07-02", "2023-07-03", "2023-07-04", "2023-07-05"]
}

df = pd.DataFrame(data)
df
```

#Part B - Identify Data Types

```python
df.info()
```

#Part 03 - Convert Data Types

```python
df['Enrolled_Date'] = pd.to_datetime(df['Enrolled_Date'], errors='coerce')
df.info()
```

```python
df['Eligibility_Status'] = np.where(df['Attendence'] >= 80, True, False)
df.info()
df
```

```python
data_1 = {"salary" : ["50000", "60000", "70000"]}
data_2 = {"Active" : ["True", "False", "True"]}
data_3 = {"Join Date" : ["2023-07-01", "2023-07-02", "2023-07-03"]}
data_4 = {"Join_Date" : [45023, 45024]}
data_5 = {"Marks" : [85, "90", "NaN", 75]}
data_6 = {"Grade": ["A", "B", "C", "A"]}

df_1 = pd.DataFrame(data_1)
df_2 = pd.DataFrame(data_2)
df_3 = pd.DataFrame(data_3)
df_4 = pd.DataFrame(data_4)
df_5 = pd.DataFrame(data_5)
df_6 = pd.DataFrame(data_6)
```

```python
df_1.dtypes
df_1['salary'] = df_1['salary'].astype(int)
df_1
```

```python
df_2.dtypes
bool_map = {
    'True' : True,
    'False' : False,
}
df_2['Active'] = df_2['Active'].map(bool_map)
df_2
```

```python
df_3.dtypes
df_3['Join Date'] = pd.to_datetime(df_3['Join Date'], errors='coerce')
df_3
```

```python
df_4['Join_Date'] = pd.to_datetime(df_4['Join_Date'], unit='D', origin='1899-12-30')
df_4
```

```python
df_5['Marks'] = pd.to_numeric(df_5['Marks'], errors='coerce')
df_5
```

```python
df_6['Grade'] = df_6['Grade'].astype('category')
df_6
```

#Part D - Work with Categorical Variables

```python
unique_grades = df['Final_Grade'].unique()
print(unique_grades)
```

```python
df['Final_Grade'] = df['Final_Grade'].astype('category')
df.dtypes
```

#Part E - Encoding Categorical Data

```python
df['Grade_Ranked'] = df['Final_Grade'].cat.codes
df
```

```python
df_encoded = pd.get_dummies(df, columns=['Final_Grade'])
df_encoded
```
